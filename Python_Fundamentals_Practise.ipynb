## 📌 Section 1 — Python Syntax, Variables & Data Types

### ✅ Example 1.1 — Variable Assignment & Multiple Data Types

Python is **dynamically typed** — you don't need to declare the type explicitly.  
Use `type()` to inspect any variable's data type at runtime.

```python
# --- Basic Variable Assignment ---
student_name = "Garth"          # str
student_age  = 25               # int
gpa          = 3.85             # float
is_enrolled  = True             # bool
courses      = ["Math", "CS", "Physics"]   # list
address      = ("Mumbai", "Maharashtra")   # tuple  (immutable)
profile      = {"roll": 101, "grade": "A"} # dict
unique_tags  = {"python", "ai", "ml"}      # set

# --- Print with type info ---
variables = [
    ("student_name", student_name),
    ("student_age",  student_age),
    ("gpa",          gpa),
    ("is_enrolled",  is_enrolled),
    ("courses",      courses),
    ("address",      address),
    ("profile",      profile),
    ("unique_tags",  unique_tags),
]

print(f"{'Variable':<15} {'Value':<35} {'Type'}")
print("-" * 65)
for name, val in variables:
    print(f"{name:<15} {str(val):<35} {type(val).__name__}")
```

### ✅ Example 1.2 — Type Conversion (Casting)

Python lets you convert between types using built-in functions like `int()`, `float()`, `str()`, `list()`, `bool()`.

```python
# --- Type Conversion Examples ---

# str → int / float
price_str   = "199"
price_int   = int(price_str)
price_float = float(price_str)
print(f"String '{price_str}' → int: {price_int}, float: {price_float}")

# float → int  (truncates decimal)
score = 87.9
print(f"\nFloat {score} → int: {int(score)}  (decimal is dropped, NOT rounded)")

# int → bool
print(f"\nbool(0)  = {bool(0)},  bool(1) = {bool(1)},  bool(-5) = {bool(-5)}")

# list ↔ set  (set removes duplicates)
marks = [90, 85, 90, 78, 85, 92]
unique_marks = list(set(marks))
print(f"\nOriginal list : {marks}")
print(f"After set→list: {sorted(unique_marks)}  ← duplicates removed")

# Any value → string
age = 25
message = "My age is " + str(age)
print(f"\n{message}")
```

### ✅ Example 1.3 — Python Syntax Rules in Practice

Key Python syntax rules:  
- Indentation defines code blocks (no `{}` needed)  
- `#` for single-line comments, `"""..."""` for multi-line  
- Multiple assignment, swapping without temp variable

```python
# --- Multiple assignment in one line ---
x, y, z = 10, 20, 30
print(f"x={x}, y={y}, z={z}")

# --- Swap without a temp variable ---
a, b = 5, 99
print(f"\nBefore swap: a={a}, b={b}")
a, b = b, a
print(f"After swap : a={a}, b={b}")

# --- Chained assignment ---
p = q = r = 100
print(f"\nChained: p={p}, q={q}, r={r}")

# --- None type (represents absence of value) ---
result = None
print(f"\nresult is None: {result is None}")
print(f"Type of None  : {type(result)}")

# --- Indentation-based block ---
temperature = 38
if temperature > 37:
    status = "Fever"
    print(f"\nTemperature: {temperature}°C → Status: {status}")
else:
    print("Normal temperature")
```

## 📌 Section 2 — Numbers, Strings & Operators

### ✅ Example 2.1 — Arithmetic & Numeric Operations

Python supports `int`, `float`, and `complex` numbers.  
Key operators: `+`, `-`, `*`, `/`, `//` (floor div), `%` (modulo), `**` (power)

```python
import math

a, b = 17, 5

print("=== Arithmetic Operators ===")
print(f"{a} + {b}  = {a + b}")
print(f"{a} - {b}  = {a - b}")
print(f"{a} * {b}  = {a * b}")
print(f"{a} / {b}  = {a / b}   ← always float")
print(f"{a} // {b} = {a // b}   ← floor division (integer)")
print(f"{a} % {b}  = {a % b}   ← remainder (modulo)")
print(f"{a} ** {b} = {a ** b}  ← exponentiation")

print("\n=== math module ===")
print(f"math.sqrt(144) = {math.sqrt(144)}")
print(f"math.ceil(4.2) = {math.ceil(4.2)}")
print(f"math.floor(4.9)= {math.floor(4.9)}")
print(f"math.pi        = {math.pi:.4f}")
print(f"math.factorial(6) = {math.factorial(6)}")

print("\n=== Practical: Circle area ===")
radius = 7
area = math.pi * radius ** 2
print(f"Radius = {radius} → Area = {area:.2f}")
```

### ✅ Example 2.2 — String Operations & Methods

Strings in Python are **immutable sequences**.  
Rich built-in methods: `upper()`, `lower()`, `split()`, `join()`, `replace()`, `strip()`, `find()`, etc.

```python
sentence = "  Hello, Python Learners! Welcome to Coding.  "

print("=== Basic String Methods ===")
print(f"Original  : '{sentence}'")
print(f"strip()   : '{sentence.strip()}'")
print(f"upper()   : '{sentence.strip().upper()}'")
print(f"lower()   : '{sentence.strip().lower()}'")
print(f"title()   : '{sentence.strip().title()}'")

clean = sentence.strip()
print(f"\nreplace() : '{clean.replace('Coding', 'Python')}'")
print(f"find()    : index of 'Python' = {clean.find('Python')}")
print(f"count()   : 'o' appears {clean.count('o')} times")
print(f"startswith: {clean.startswith('Hello')}")
print(f"endswith  : {clean.endswith('Coding.')}")

print("\n=== Slicing ===")
word = "PythonProgramming"
print(f"word[0:6]  = {word[0:6]}")
print(f"word[-11:] = {word[-11:]}")
print(f"word[::2]  = {word[::2]}   ← every 2nd char")
print(f"word[::-1] = {word[::-1]}  ← reversed")

print("\n=== f-strings (formatted output) ===")
name, score = "Alice", 95.678
print(f"Student: {name:>10} | Score: {score:.2f} | Grade: {'A' if score >= 90 else 'B'}")

print("\n=== split & join ===")
csv_line = "apple,banana,cherry,mango"
fruits = csv_line.split(",")
print(f"split result : {fruits}")
print(f"join back    : {'  |  '.join(fruits)}")
```

### ✅ Example 2.3 — Comparison, Logical & Bitwise Operators

Python comparison operators return `True`/`False`.  
Logical: `and`, `or`, `not` | Bitwise: `&`, `|`, `^`, `~`, `<<`, `>>`

```python
# --- Comparison Operators ---
x, y = 15, 20
print("=== Comparison Operators ===")
print(f"{x} == {y}  → {x == y}")
print(f"{x} != {y}  → {x != y}")
print(f"{x} < {y}   → {x < y}")
print(f"{x} > {y}   → {x > y}")
print(f"{x} <= {y}  → {x <= y}")
print(f"{x} >= {y}  → {x >= y}")

# --- Logical Operators ---
age, has_id = 22, True
print("\n=== Logical Operators ===")
print(f"Can enter club (age>=18 AND has_id): {age >= 18 and has_id}")
print(f"Get discount   (age<12 OR age>60)  : {age < 12 or age > 60}")
print(f"Is minor       (NOT age>=18)       : {not age >= 18}")

# --- Membership & Identity ---
fruits = ["apple", "banana", "mango"]
print("\n=== Membership Operators ===")
print(f"'mango'  in fruits  → {'mango' in fruits}")
print(f"'grape'  in fruits  → {'grape' in fruits}")
print(f"'grape' not in fruits → {'grape' not in fruits}")

# --- Augmented Assignment ---
counter = 10
print("\n=== Augmented Assignment ===")
counter += 5;  print(f"After += 5  : {counter}")
counter -= 3;  print(f"After -= 3  : {counter}")
counter *= 2;  print(f"After *= 2  : {counter}")
counter //= 3; print(f"After //= 3 : {counter}")
counter **= 2; print(f"After **= 2 : {counter}")
```

## 📌 Section 3 — Python Functions & Modular Programming

### ✅ Example 3.1 — Function Types: Basic, Default Args, *args, **kwargs

Functions are defined with `def`. Python supports:  
- **Positional** args → required in order  
- **Default** args → optional with fallback values  
- **`*args`** → variable positional arguments (tuple)  
- **`**kwargs`** → variable keyword arguments (dict)

```python
# --- 1. Basic function ---
def greet(name):
    """Returns a greeting message."""
    return f"Hello, {name}! 👋"

print(greet("Garth"))

# --- 2. Default arguments ---
def calculate_power(base, exponent=2):
    """Calculates base raised to exponent (default: square)."""
    return base ** exponent

print(f"\ncalculate_power(4)    = {calculate_power(4)}     ← uses default exp=2")
print(f"calculate_power(2, 8) = {calculate_power(2, 8)}  ← custom exponent")

# --- 3. *args — accepts any number of positional args ---
def total_marks(*marks):
    """Sums any number of subject marks."""
    total = sum(marks)
    avg   = total / len(marks)
    return total, round(avg, 2)

total, avg = total_marks(85, 90, 78, 92, 88)
print(f"\nTotal: {total}, Average: {avg}")

# --- 4. **kwargs — accepts keyword arguments ---
def create_student_profile(**info):
    """Builds a student profile from keyword arguments."""
    print("\n=== Student Profile ===")
    for key, value in info.items():
        print(f"  {key.capitalize():<12}: {value}")

create_student_profile(name="Alice", age=21, course="CS", gpa=3.9, city="Pune")
```

### ✅ Example 3.2 — Lambda Functions & Higher-Order Functions

`lambda` creates small anonymous functions in one line.  
Higher-order functions (`map`, `filter`, `sorted`) accept functions as arguments.

```python
# --- Lambda: basic ---
square    = lambda x: x ** 2
add       = lambda x, y: x + y
is_even   = lambda n: n % 2 == 0

print(f"square(7)    = {square(7)}")
print(f"add(3, 4)    = {add(3, 4)}")
print(f"is_even(10)  = {is_even(10)}")

# --- map() — applies function to every element ---
prices = [100, 250, 80, 400, 175]
discounted = list(map(lambda p: round(p * 0.85, 2), prices))
print(f"\nOriginal prices  : {prices}")
print(f"After 15% discount: {discounted}")

# --- filter() — keeps elements where function returns True ---
numbers = list(range(1, 21))
evens   = list(filter(lambda n: n % 2 == 0, numbers))
print(f"\nEven numbers in 1–20: {evens}")

# --- sorted() with key ---
students = [
    {"name": "Bob",   "gpa": 3.5},
    {"name": "Alice", "gpa": 3.9},
    {"name": "Eve",   "gpa": 3.7},
    {"name": "Dan",   "gpa": 3.2},
]
ranked = sorted(students, key=lambda s: s["gpa"], reverse=True)
print("\n=== Students ranked by GPA ===")
for i, s in enumerate(ranked, 1):
    print(f"  {i}. {s['name']:<8} GPA: {s['gpa']}")
```

### ✅ Example 3.3 — Modular Programming: Scope, Recursion & Docstrings

- **Scope**: `local` vs `global` variables  
- **Recursion**: function calling itself (needs a base case)  
- **Docstrings**: document functions with `help()`

```python
# --- Scope: local vs global ---
company = "TechCorp"      # global

def show_employee(name):
    department = "Engineering"  # local
    print(f"{name} works at {company} in {department}")

show_employee("Garth")
# print(department)  # ← would raise NameError — local scope!

# --- Recursion: Factorial ---
def factorial(n):
    """
    Computes n! recursively.

    Args:
        n (int): Non-negative integer.

    Returns:
        int: Factorial of n.
    """
    if n == 0 or n == 1:   # base case
        return 1
    return n * factorial(n - 1)   # recursive call

print("\n=== Factorial using Recursion ===")
for n in range(1, 8):
    print(f"  {n}! = {factorial(n)}")

# --- Recursion: Fibonacci ---
def fibonacci(n):
    """Returns nth Fibonacci number."""
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

print("\n=== Fibonacci Sequence (first 10) ===")
fib_seq = [fibonacci(i) for i in range(10)]
print(f"  {fib_seq}")

# --- help() reads the docstring ---
print("\n=== Docstring via help() ===")
help(factorial)
```

## 📌 Section 4 — Object-Oriented Programming (OOP)

### ✅ Example 4.1 — Classes, Objects & `__init__` (Constructor)

A **class** is a blueprint; an **object** is an instance.  
`__init__` is the constructor — called automatically when you create an object.  
`self` refers to the current object instance.

```python
class BankAccount:
    """Represents a simple bank account."""

    bank_name = "PyBank"    # class attribute (shared by all instances)

    def __init__(self, owner, balance=0):
        """Constructor: initialise owner and balance."""
        self.owner   = owner       # instance attribute
        self.balance = balance
        self.transactions = []

    def deposit(self, amount):
        """Add money to account."""
        self.balance += amount
        self.transactions.append(f"+{amount}")
        print(f"  Deposited ₹{amount}. New balance: ₹{self.balance}")

    def withdraw(self, amount):
        """Withdraw money from account."""
        if amount > self.balance:
            print(f"  ❌ Insufficient funds! Balance: ₹{self.balance}")
        else:
            self.balance -= amount
            self.transactions.append(f"-{amount}")
            print(f"  Withdrew ₹{amount}. New balance: ₹{self.balance}")

    def statement(self):
        """Print account summary."""
        print(f"\n  {'='*35}")
        print(f"  Bank        : {BankAccount.bank_name}")
        print(f"  Account     : {self.owner}")
        print(f"  Transactions: {', '.join(self.transactions)}")
        print(f"  Balance     : ₹{self.balance}")
        print(f"  {'='*35}")

    def __str__(self):
        """String representation of the object."""
        return f"BankAccount(owner='{self.owner}', balance=₹{self.balance})"


# --- Create objects (instances) ---
acc1 = BankAccount("Garth", 5000)
acc2 = BankAccount("Alice")

print("=== Account Operations ===")
acc1.deposit(2000)
acc1.withdraw(1500)
acc1.withdraw(10000)    # should fail
acc1.statement()

print(f"\nString repr: {acc1}")
print(f"Same bank  : {acc2.bank_name}")
```

### ✅ Example 4.2 — Inheritance & Method Overriding

**Inheritance** lets a child class reuse and extend a parent class.  
- `super()` calls the parent class constructor/method  
- **Method Overriding**: child redefines a parent method

```python
# --- Parent (Base) Class ---
class Animal:
    def __init__(self, name, species):
        self.name    = name
        self.species = species

    def speak(self):
        return f"{self.name} makes a sound."

    def info(self):
        return f"{self.name} is a {self.species}"

    def __str__(self):
        return f"Animal({self.name}, {self.species})"


# --- Child Class: inherits Animal ---
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, species="Dog")   # call parent __init__
        self.breed = breed

    def speak(self):           # Method Override
        return f"{self.name} says: Woof! 🐕"

    def fetch(self, item):
        return f"{self.name} fetched the {item}!"


class Cat(Animal):
    def __init__(self, name, indoor=True):
        super().__init__(name, species="Cat")
        self.indoor = indoor

    def speak(self):           # Method Override
        return f"{self.name} says: Meow! 🐈"

    def info(self):            # Extending parent info()
        base = super().info()
        return f"{base} ({'Indoor' if self.indoor else 'Outdoor'})"


class Bird(Animal):
    def __init__(self, name, can_fly=True):
        super().__init__(name, species="Bird")
        self.can_fly = can_fly

    def speak(self):           # Method Override
        return f"{self.name} says: Tweet! 🐦"


# --- Instantiate objects ---
dog  = Dog("Bruno", "Labrador")
cat  = Cat("Whiskers", indoor=True)
bird = Bird("Tweety")

print("=== Animal Sounds ===")
for animal in [dog, cat, bird]:
    print(f"  {animal.speak()}")

print("\n=== Info ===")
print(f"  {dog.info()}  |  Breed: {dog.breed}")
print(f"  {cat.info()}")

print(f"\n  {dog.fetch('ball')}")

# --- isinstance check ---
print("\n=== isinstance checks ===")
print(f"  dog is Animal? {isinstance(dog, Animal)}")
print(f"  dog is Dog?    {isinstance(dog, Dog)}")
print(f"  dog is Cat?    {isinstance(dog, Cat)}")
```

### ✅ Example 4.3 — Polymorphism & Encapsulation

**Polymorphism**: same method name, different behaviour across classes.  
**Encapsulation**: hiding internal data using access modifiers:  
- `public` → `self.name` (accessible anywhere)  
- `_protected` → `self._salary` (convention: internal use)  
- `__private` → `self.__password` (name-mangled, not directly accessible)

```python
# ============================================================
#  POLYMORPHISM
# ============================================================

class Shape:
    def area(self):
        raise NotImplementedError("Subclasses must implement area()")

    def describe(self):
        return f"I am a {self.__class__.__name__} with area = {self.area():.2f}"


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):              # Polymorphic method
        import math
        return math.pi * self.radius ** 2


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width  = width
        self.height = height

    def area(self):              # Polymorphic method
        return self.width * self.height


class Triangle(Shape):
    def __init__(self, base, height):
        self.base   = base
        self.height = height

    def area(self):              # Polymorphic method
        return 0.5 * self.base * self.height


shapes = [Circle(5), Rectangle(4, 6), Triangle(3, 8)]

print("=== Polymorphism — Shape Areas ===")
for shape in shapes:
    print(f"  {shape.describe()}")


# ============================================================
#  ENCAPSULATION
# ============================================================

class Employee:
    def __init__(self, name, salary, password):
        self.name       = name           # public
        self._salary    = salary         # protected (convention)
        self.__password = password       # private (name-mangled)

    # Getter for private attribute
    def get_salary(self):
        return self._salary

    # Setter with validation
    def set_salary(self, new_salary):
        if new_salary < 0:
            print("  ❌ Salary cannot be negative!")
        else:
            self._salary = new_salary
            print(f"  ✅ Salary updated to ₹{self._salary}")

    def check_password(self, pwd):
        return self.__password == pwd

    def __str__(self):
        return f"Employee(name={self.name}, salary=₹{self._salary})"


emp = Employee("Garth", 75000, "secret123")

print("\n=== Encapsulation Demo ===")
print(f"  Name   (public)   : {emp.name}")
print(f"  Salary (protected): {emp.get_salary()}")
print(f"  Password check    : {emp.check_password('secret123')}")

emp.set_salary(85000)
emp.set_salary(-500)    # should be rejected

# Accessing private attribute directly fails
try:
    print(emp.__password)
except AttributeError as e:
    print(f"  Direct access error: {e}")

# Name mangling — still accessible via _ClassName__attr
print(f"  Via name-mangling : {emp._Employee__password}")
print(f"\n  {emp}")
```

### ✅ Example 4.4 — Full OOP Mini Project: Student Management System

Combines **all OOP concepts** in one cohesive example:  
class + inheritance + polymorphism + encapsulation + dunder methods

```python
class Person:
    """Base class representing a person."""

    def __init__(self, name, age):
        self.name = name
        self._age  = age     # protected

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if value < 0 or value > 120:
            raise ValueError("Invalid age!")
        self._age = value

    def introduce(self):
        return f"Hi, I'm {self.name}, age {self._age}."


class Student(Person):
    """Inherits Person. Adds grade tracking."""

    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id
        self.__grades   = {}    # private — subject: score

    def add_grade(self, subject, score):
        if 0 <= score <= 100:
            self.__grades[subject] = score
        else:
            print(f"  ❌ Invalid score {score} for {subject}")

    def average(self):
        if not self.__grades:
            return 0
        return round(sum(self.__grades.values()) / len(self.__grades), 2)

    def grade_letter(self):
        avg = self.average()
        if avg >= 90: return "A"
        elif avg >= 80: return "B"
        elif avg >= 70: return "C"
        elif avg >= 60: return "D"
        else: return "F"

    def report(self):
        print(f"\n  {'─'*38}")
        print(f"  Student Report Card")
        print(f"  {'─'*38}")
        print(f"  Name       : {self.name}")
        print(f"  ID         : {self.student_id}")
        print(f"  Age        : {self.age}")
        print(f"  Subjects   :")
        for sub, score in self.__grades.items():
            bar = "█" * (score // 10)
            print(f"    {sub:<12}: {score:>3}  {bar}")
        print(f"  Average    : {self.average()}")
        print(f"  Grade      : {self.grade_letter()}")
        print(f"  {'─'*38}")

    def introduce(self):    # Polymorphic override
        base = super().introduce()
        return f"{base} I'm a student (ID: {self.student_id})."

    def __str__(self):
        return f"Student({self.name}, ID={self.student_id}, Avg={self.average()})"

    def __repr__(self):
        return self.__str__()


class GraduateStudent(Student):
    """Multi-level inheritance: GradStudent → Student → Person."""

    def __init__(self, name, age, student_id, thesis_topic):
        super().__init__(name, age, student_id)
        self.thesis_topic = thesis_topic

    def introduce(self):    # Further polymorphic override
        base = super().introduce()
        return f"{base} My thesis: '{self.thesis_topic}'."


# --- Run the system ---
s1 = Student("Garth",  23, "S001")
s2 = Student("Alice",  21, "S002")
s3 = GraduateStudent("Bob", 26, "G001", "AI in Healthcare")

s1.add_grade("Math",    92)
s1.add_grade("Python",  88)
s1.add_grade("Physics", 79)
s1.add_grade("English", 95)

s2.add_grade("Math",    75)
s2.add_grade("Python",  82)
s2.add_grade("Physics", 68)

print("=== Introductions (Polymorphism) ===")
for person in [s1, s2, s3]:
    print(f"  {person.introduce()}")

s1.report()
s2.report()

print("\n=== Ranked Leaderboard ===")
roster = sorted([s1, s2], key=lambda s: s.average(), reverse=True)
for rank, stu in enumerate(roster, 1):
    print(f"  #{rank} {stu}")
```
